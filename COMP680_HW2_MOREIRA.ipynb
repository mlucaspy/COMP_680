**Question 11**: 
Here are some common random variables (distributions) with their mean and variance listed. Pick one distribution and derive the mean and the variance. You can look up the PMF or PDF for these distributions! 

$$
\begin{array}{lll}
\text{Random Variable / Distribution} & \text{Mean} & \text{Variance}           \\
\hline
\text{Bernoulli}(p)          & p             & p(1-p)         \\
\text{Binomial}(n, p)        & np            & np(1-p)        \\
\text{Poisson}(\lambda)      & \lambda       & \lambda        \\
\text{Uniform}(a, b)         & (a + b) / 2   & (b - a)^2 / 12 \\
\text{Normal}(\mu, \sigma^2) & \mu           & \sigma^2       \\
\text{Exponential}(\beta)    & \beta         & \beta^2        \\
\text{Gamma}(\alpha, \beta)  & \alpha \beta  & \alpha \beta^2 \\
\text{Beta}(\alpha, \beta)   & \alpha / (\alpha + \beta) & \alpha \beta / ((\alpha + \beta)^2 (\alpha + \beta + 1)) \\
\end{array}
$$

**Solution**:

### Mean of Poisson Distribution ###

The poisson PMF is given by:

$$
f(k; \lambda)=\frac{\lambda^{k} e^{-\lambda}}{k !}
$$

Where $k$ is the number of events, and $\lambda$ is the average number of events per interval. $E[X]$ is the expected value of the random variable $X$.

$k$ is a discrete random variable, so the mean is given by:

$$
E[X] =\sum_{k=0}^{\infty} k \frac{\lambda^{k} e^{-\lambda}}{k !}
$$

We can rewrite the sum as, as k=0 is the first term of the sum ad does not change the result:
$$
=\sum_{k=1}^{\infty} k \frac{\lambda^{k} e^{-\lambda}}{k !}
$$

We simplify the last term of the factorial:
$$
=\sum_{k=1}^{\infty} \frac{\lambda^{k} e^{-\lambda}}{(k-1) !}
$$

We can rewrite the sum as, taking lambda out of the sum:
$$
=\lambda \sum_{k=1}^{\infty} \frac{\lambda^{k-1} e^{-\lambda}}{(k-1) !}
$$

This expression is the same as the poisson PMF, but with k-1 instead of k. We can rewrite the sum as:
$$
=\lambda \sum_{k=0}^{\infty} \frac{\lambda^{k} e^{-\lambda}}{k !}
$$

The sum is the same as the poisson PMF, so the sum is equal to 1:
$$
=\lambda
$$


### Variance of Poisson Distribution ###

The variance is given by:

$$
Var[X] = E[X^2] - E[X]^2
$$

We can rewrite the variance as:
$$
Var[X] = E[X^2] - \lambda^2
$$

We can rewrite the first term of the variance as:
$$
E[X^2] = \sum_{k=0}^{\infty} k^2 \frac{\lambda^{k} e^{-\lambda}}{k !}
$$

We can rewrite the sum as, as k=0 is the first term of the sum ad does not change the result:
$$
=\sum_{k=1}^{\infty} k^2 \frac{\lambda^{k} e^{-\lambda}}{k !}
$$

We simplify the last term of the factorial:
$$
=\sum_{k=1}^{\infty} k \frac{\lambda^{k} e^{-\lambda}}{(k-1) !}
$$

We can rewrite the sum as, taking lambda out of the sum:
$$
=\lambda \sum_{k=1}^{\infty} k \frac{\lambda^{k-1} e^{-\lambda}}{(k-1) !}\\
= \lambda \sum_{k=1}^{\infty} (k -1 +1) \frac{\lambda^{k-1} e^{-\lambda}}{(k-1) !} \\
= \lambda \sum_{k=1}^{\infty} (k-1) \frac{\lambda^{k-1} e^{-\lambda}}{(k-1) !} + \lambda \sum_{k=1}^{\infty} \frac{\lambda^{k-1} e^{-\lambda}}{(k-1) !}\\
= \lambda \lambda \sum_{k=2}^{\infty} \frac{\lambda^{k-2} e^{-\lambda}}{(k-1) !} + \lambda  + \lambda\\
= \lambda^2 + \lambda
$$

We can rewrite the variance as:
$$
Var[X] = E[X^2] - \lambda^2\\
= \lambda^2 + \lambda - \lambda^2\\
= \lambda
$$

**Question 12**: If $X_1, X_2, \cdots X_n$ are IID (independent and identically distributed) from a distribution with mean $\mu$ and variance $\sigma^2$. Calculate the mean and variance for the sample mean $\bar{X}_n = \sum_{i=1}^n X_i/ n$.

**Solution**: 

### Explaination of what $X_1, X_2, \cdots X_n$ are:

$X_1, X_2, \cdots X_n$ are IID (independent and identically distributed) random variables. This means that each random variable is independent of each other, and that each random variable has the same distribution.

If each $X_i$ is a random variable, then the sample mean $\bar{X}_n$ is also a random variable.

### Calculate the mean and variance for the sample mean $\bar{X}_n = \sum_{i=1}^n X_i/ n$:

The mean of the sample mean is given by:

$$
E[\bar{X}_n] = E[\sum_{i=1}^n X_i/ n]\\
= \sum_{i=1}^n E[X_i/ n]\\
= \sum_{i=1}^n E[X_i]/ n\\
= \sum_{i=1}^n \mu/ n\\
= \mu
$$

The variance of the sample mean is given by:

$$
Var[\bar{X}_n] = Var[\sum_{i=1}^n X_i/ n]\\
= \sum_{i=1}^n Var[X_i/ n]\\
= \sum_{i=1}^n Var[X_i]/ n^2\\
= \sum_{i=1}^n \sigma^2/ n^2\\
= \sigma^2/ n
$$

**Question 13**: Let $X \sim \text{Uniform}(0, 1)$.  Let $0 < a < b < 1$.  Let

$$
Y = \begin{cases}
1 &\text{if } 0 < x < b \\
0 &\text{otherwise}
\end{cases}
$$

and let

$$
Z = \begin{cases}
1 &\text{if } a < x < 1 \\
0 &\text{otherwise}
\end{cases}
$$

- Are $Y$ and $Z$ independent?  Why or Why not?
- Find $\mathbb{P}(Y | Z)$, then use this conditional distribution to find $\mathbb{E}(Y | Z)$.  

**Solution**:

### Are $Y$ and $Z$ independent?  Why or Why not? Write this in terms of $a$ and $b$.

**Random Variable X**  
| 0 ------ a ------ b ------ 1 ------ >

As the probability of $Y$ depends on $b$, and the probability of $Z$ depends on $a$, and $a$ depends on $b$. So de probability of $Z$ depends on $b$. So $Y$ and $Z$ are not independent.


### Find $\mathbb{P}(Y | Z)$, then use this conditional distribution to find $\mathbb{E}(Y | Z)$.

The probability of $Y$ given $Z$ is given by:

$$
\mathbb{P}(Y | Z) = \frac{\mathbb{P}(Y \cap Z)}{\mathbb{P}(Z)}
$$

We can interpret geometricaly. We can see that $Y$ covers a region from $0$ to $b$, and $Z$ covers a region from $a$ to $1$. So the intersection of $Y$ and $Z$ is the region from $a$ to $b$. So the probability of $Y$ given $Z$ is given by:

$$
\mathbb{P}(Y | Z) = \frac{\mathbb{P}(a < x < b)}{\mathbb{P}(a < x < 1)}
$$

As we know that $X$ has a uniform distribution, and taking in consideration the geometrical interpretation, we can rewrite the probability of $Y$ given $Z$ as:

$$
\mathbb{P}(Y | Z) = \frac{b - a}{1 - a}
$$

### Expected value of $Y$ given $Z$:

The expected value of $Y$ given $Z$ is given by:

$$
\mathbb{E}(Y | Z) = \sum_{y} y \mathbb{P}(Y = y | Z)
$$

We can rewrite the expected value of $Y$ given $Z$ as:

$$
\mathbb{E}(Y | Z) = 1 \mathbb{P}(Y = 1 | Z) + 0 \mathbb{P}(Y = 0 | Z) \\
\mathbb{E}(Y | Z) = 1 \frac{b - a}{1 - a} + 0 \frac{1 - b}{1 - a}
$$

So the expected value of $Y$ given $Z$ is:

$$
\mathbb{E}(Y | Z) = \frac{b - a}{1 - a}
$$

**Question 14**: Suppose we generate a random variable $X$ in the following way:
- First we flip a fair coin. 
- If the coin is heads, take $X$ to have a $\text{Uniform}(0, 1)$ distribution.  
- If the coin is tails, take $X$ to have a $\text{Uniform}(3, 4)$ distribution.

Find $\mathbb{E}(X)$: the expectation of $X$.

**Solution**:

The expected value of $X$ is given by:

$$
\mathbb{E}(X) = \mathbb{E}(X | \text{heads}) \mathbb{P}(\text{heads}) + \mathbb{E}(X | \text{tails}) \mathbb{P}(\text{tails})
$$

We can rewrite the expected value of $X$ as:
$$
\mathbb{E}(X) = \mathbb{E}(X | \text{heads}) \mathbb{P}(\text{heads}) + \mathbb{E}(X | \text{tails}) \mathbb{P}(\text{tails})\\
= \mathbb{E}(\text{Uniform}(0, 1)) \frac{1}{2} + \mathbb{E}(\text{Uniform}(3, 4)) \frac{1}{2}
$$

The expected value of a uniform distribution is given by:
$$
\mathbb{E}(\text{Uniform}(a, b)) = \frac{a + b}{2}
$$

So the expected value of $X$ is given by:
$$
\mathbb{E}(X) = \mathbb{E}(\text{Uniform}(0, 1)) \frac{1}{2} + \mathbb{E}(\text{Uniform}(3, 4)) \frac{1}{2}\\
= \frac{0 + 1}{2} \frac{1}{2} + \frac{3 + 4}{2} \frac{1}{2}\\
= \frac{1}{4} + \frac{7}{4}\\
= 2
$$

**Question 15**: Let $Y_1, Y_2, \dots$ be independent random variables such that $\mathbb{P}(Y_i = 1) = \mathbb{P}(Y_i = -1) = 1/2$.  Let $X_n = \sum_{i=1}^n Y_i$.  Think of $Y_i = 1$ as "the stock price increased by one dollar" $Y_i = -1$ as "the stock price decreased by one dollar" and $X_n$ as the value of the stock on day $n$.

Find 
-  $\mathbb{E}(X_n)$: the expectation of $X_n$ 
- $\mathbb{V}(X_n)$: the variance of $X_n$.

**Solution**:

1. **Expectation $\mathbb{E}(X_n)$**:
   The expectation of $X_n$ is given by the sum of the expectations of the individual random variables $Y_i$. Since $Y_i$ is either 1 or -1 with equal probability, its expectation is:
   
   $\mathbb{E}(Y_i) = 1 \cdot \frac{1}{2} + (-1) \cdot \frac{1}{2} = 0$

   Therefore, for $X_n$, which is the sum of $n$ independent $Y_i$ variables:
   
   $\mathbb{E}(X_n) = \sum_{i=1}^n \mathbb{E}(Y_i) = \sum_{i=1}^n 0 = 0$

   So, the expectation of $X_n$ is always 0.

2. **Variance $\mathbb{V}(X_n)$**:

   $$
   \mathbb{V}(Y_i) = \mathbb{E}(Y_i^2) - (\mathbb{E}(Y_i))^2 \\
   = 1^2 \times \frac{1}{2} + (-1)^2 \times \frac{1}{2} - 0^2 \\
   = \frac{1}{2} + \frac{1}{2} = 1
   $$

   To find the variance of $X_n$, we can use the fact that the $Y_i$ variables are independent. The variance of the sum of independent random variables is equal to the sum of their variances. Since each $Y_i$ has variance $\mathbb{V}(Y_i) = 1$, we can calculate the variance of $X_n$:
   
   $\mathbb{V}(X_n) = \sum_{i=1}^n \mathbb{V}(Y_i) = \sum_{i=1}^n 1 = n$

   So, the variance of $X_n$ is $n$.

**Just for Fun: Simulating the Stock Market**  
Simulate $X_n$ and plot $X_n$ versus $n$ for $n = 1, 2, \dots, 1,000$.  Repeat the whole simulation several times.  

```python
# total number of days N
N = 1000
# how many times to repeat the simulation
B = 10

# the shape of Y should be (20, 1000)
Y = 2 * np.random.binomial(n=1, p=1/2, size=(B, N)) - 1

# np.cumsum can calculate X_n = sum_{i=1}^n Y_i
# the shape of X is the same of Y
# X is the observed daily price for B stocks for N days
X = np.cumsum(Y, axis=1)
```

```python
# define the days as a numpy array
t = np.arange(1, N + 1)

# specify the figure size 
plt.figure(figsize=(12, 8))

# 1.96 is a magic number
z = 1.96
# the following 3 lines created a shaded range to predict
# the price change of the stock: the 95% confidence interval
plt.plot(t, z * np.sqrt(t), color='red')
plt.plot(t, -z * np.sqrt(t), color='red')
plt.fill_between(t, z * np.sqrt(t), -z * np.sqrt(t), color='red', alpha=0.1)

# plot each of stock price trajectory simulated
for b in range(B):
    plt.plot(t, X[b])

plt.title("Simulated Daily Stock Prices")
plt.xlabel("Number of Days")
plt.ylabel("Price change in $")
plt.show()
```
